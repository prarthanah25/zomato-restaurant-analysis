```python
df = df.rename(columns={
    "approx_cost(for two people)": "cost_for_two",
    "listed_in(type)": "restaurant_type",
    "listed_in(city)": "city"
})
```

```python
## Renaming Columns

Some column names were long and difficult to work with. They were renamed to shorter, more meaningful names to improve readability and make coding easier.
```

```python
df["rate"].head(10)
```

```python
df["rate"].unique()
```

```python
df["rate"] = df["rate"].replace("NEW", np.nan)
df["rate"] = df["rate"].replace("-", np.nan)
```

```python
df["rate"] = df["rate"].str.replace("/5", "", regex=False)
```

```python
df["rate"] = df["rate"].astype(float)
```
